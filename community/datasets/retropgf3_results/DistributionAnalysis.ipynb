```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("csv/RetroPGF3 Results - ConsolidatedData.csv", index_col=0)
df.iloc[0]
```

```python
fig, ax = plt.subplots(facecolor='white', figsize=(15,5))
df['Result: Received OP'].plot(kind='area', lw=2)
ax.set_xlim(0,500)
ax.set_ylim(0,700_000)
ax.set_xticks([])
ax.set_ylabel("Received OP")
```

```python
fig, ax = plt.subplots(facecolor='white', figsize=(15,15))
sns.scatterplot(
    data=df, 
    x='Result: # Ballots', 
    y='Result: Received OP', 
    hue='Meta: Applicant Type', 
    alpha=.25, 
    palette=['blue', 'red'],
)
ax.set_xlim(0,120)
ax.set_ylim(0,700_000)
```

```python
fig, ax = plt.subplots(facecolor='white', figsize=(15,5))

dff = df[(df['Result: Received OP'] > 0) & (df['OSO: Avg Monthly Active Devs Last 6 Months'] >= 1)]
op_per_dev = (
    (dff['Result: Received OP'] / dff['OSO: Avg Monthly Active Devs Last 6 Months'])
    .sort_values(ascending=False)
    .dropna()
    .rename("Received OP per Active Dev")
)
op_per_dev.plot(kind='area', lw=2)
ax.set_xlim(0,len(op_per_dev))
ax.set_xticks([])
ax.set_ylabel("Received OP per Active Dev")
op_per_dev.describe()
```

```python
fig, ax = plt.subplots(facecolor='white', figsize=(15,15))
sns.scatterplot(
    data=dff, 
    x='OSO: Avg Monthly Active Devs Last 6 Months', 
    y=op_per_dev, 
    size='Result: Received OP', 
    alpha=.5, 
    legend=False,
    sizes=(10,1_000)
)
```

```python
fig, ax = plt.subplots(facecolor='white', figsize=(15,15))
sns.scatterplot(
    data=df, 
    x='Result: # Ballots', 
    y='Result: Received OP', 
    hue='OSO: Has Profile', 
    alpha=.5
)
ax.set_xlim(0,120)
ax.set_ylim(0,700_000)
```
